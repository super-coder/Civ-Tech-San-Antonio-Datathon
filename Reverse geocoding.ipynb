```python
import numpy as np
coordinates = pd.DataFrame(data, columns=('b.lat','b.long', 'b.u_id'))
```

```python
coordinates
```

```python
coordinates=coordinates.drop_duplicates()
```

```python
coordinates=coordinates.as_matrix()
coordinates
```

```python
len(coordinates)
```

```python
import numpy as np
from bs4 import BeautifulSoup
import urllib.request
import json


address=[]
for i in range(len(coordinates)):
    qpage = 'https://maps.googleapis.com/maps/api/geocode/json?latlng='+str(coordinates[i][0])+','+str(coordinates[i][1])+'&key=AIzaSyAnNY3f-EsQCoPkZ2n5Uj71dQ9rcQMhmno'
    page = urllib.request.urlopen(qpage)
    data = page.read().decode('utf-8')
    js = json.loads(data)
    address.append(js['results'][1]['formatted_address'])
```
