```python
def map_addnum(x):
    try:
        address = usaddress.parse(x.upper().replace(',','').replace('.',''))
        address=dict(address)
        address = {v: k for k, v in address.items()}
        AddressNumber=address['AddressNumber']
        return AddressNumber
    except: None
    return ' '
```

```python
def st_dir(x):
    try:
        address = usaddress.parse(x.upper().replace(',','').replace('.',''))
        address=dict(address)
        address = {v: k for k, v in address.items()}
        StreetDirection=address['StreetNamePreDirectional']
        return StreetDirection
    except: None
    return ' '
```

```python
def st_name(x):
    try:
        address = usaddress.parse(x.upper().replace(',','').replace('.',''))
        address=dict(address)
        address = {v: k for k, v in address.items()}
        StreetName=address['StreetName']
        return StreetName
    except: None
    return ' '
```

```python
def st_post_num(x):
    try:
        address = usaddress.parse(x.upper().replace(',','').replace('.',''))
        address=dict(address)
        address = {v: k for k, v in address.items()}
        StreetNamePostType=address['StreetNamePostType']
        return StreetNamePostType
    except: None
    return ' '
```

```python
def apt_num(x):
    try:
        address = usaddress.parse(x.upper().replace(',','').replace('.',''))
        address=dict(address)
        address = {v: k for k, v in address.items()}
        OccupancyIdentifier=address['OccupancyIdentifier']
        return OccupancyIdentifier
    except: None
    return ' '
```

```python
add['AddressNumber'] = add['Property Address'].apply(lambda x: map_addnum(x))
add['StreetDirection'] = add['Property Address'].apply(lambda x: st_dir(x))
add['StreetName'] = add['Property Address'].apply(lambda x: st_name(x))
add['StreetPostName'] = add['Property Address'].apply(lambda x: st_post_num(x))
add['OccupancyIdentifier'] = add['Property Address'].apply(lambda x: apt_num(x))
add
```
