```python
import pandas as pd
```

```python
main = pd.read_csv('feb17_consumption.csv')
main = main.drop(columns=['Unnamed: 0'])
main = main.set_index('u_id')
main
```

```python
prices = pd.read_csv("Price_2017_CPS.csv")
prices['hour'] = pd.to_numeric(prices.HOUR_ENDING, errors=('coerce'))//100 - 1
prices['minute'] = (pd.to_numeric(prices.INTERVAL_ENDING, errors='coerce') - pd.to_numeric(prices.HOUR_ENDING, errors='coerce') + 100)%100
prices = prices.dropna()
prices['timestamp'] = prices.apply(lambda row: pd.Timestamp('{0} {1}:{2}'.format(row.OPERATION_DATE, row.hour, row.minute)), axis=1)
prices = prices.drop(columns=['OPERATION_DATE', 'HOUR_ENDING', 'INTERVAL_ENDING', 'TOUTYPE', 'MARKET', 'ZONE'])
prices['price'] = pd.to_numeric(prices['PRICE'])
prices = prices.drop(columns=['PRICE'])
prices
```

```python
mar17 = pd.read_csv("march_2017.csv")
mar17['timestamp'] = mar17['a.local_interval_time']
mar17 = mar17.drop(columns=['a.local_interval_time'])
mar17['timestamp'] = pd.to_datetime(mar17['timestamp'])
mar17['lp_value'] = pd.to_numeric(mar17['a.lp_value'])
mar17['Latitude'] = pd.to_numeric(mar17['b.lat'])
mar17['Longitude'] = pd.to_numeric(mar17['b.long'])
mar17['u_id'] = pd.to_numeric(mar17['b.u_id'])
mar17 = mar17.drop(columns=['b.lat', 'b.long', 'a.lp_value', 'b.u_id'])
mar17_prices = pd.merge(prices, mar17, on='timestamp', how='inner')
mar17_prices = mar17_prices.dropna()
mar17_prices = mar17_prices.drop(columns=['hour', 'minute'])
mar17_prices['Mar17'] = mar17_prices['price']*mar17_prices['lp_value']/100
consumption = mar17_prices.groupby('u_id').sum()
consumption = consumption.drop(columns=['price', 'lp_value', 'Longitude', 'Latitude'])
del mar17, mar17_prices
```

```python
data = pd.merge(main, consumption, on='u_id', how='left')
data
```
