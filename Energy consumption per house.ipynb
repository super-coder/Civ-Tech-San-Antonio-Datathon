```python
import pandas as pd
from datetime import time
```

```python
prices = pd.read_csv("Price_2017_CPS.csv")
```

```python
prices
```

```python
prices['hour'] = pd.to_numeric(prices.HOUR_ENDING, errors=('coerce'))//100 - 1
prices['minute'] = (pd.to_numeric(prices.INTERVAL_ENDING, errors='coerce') - pd.to_numeric(prices.HOUR_ENDING, errors='coerce') + 100)%100
prices = prices.dropna()
prices
```

```python
from pytz import UTC
```

```python
prices['timestamp'] = prices.apply(lambda row: pd.Timestamp('{0} {1}:{2}'.format(row.OPERATION_DATE, row.hour, row.minute)), axis=1)
prices
```

```python
prices = prices.drop(columns=['OPERATION_DATE', 'HOUR_ENDING', 'INTERVAL_ENDING', 'TOUTYPE', 'MARKET', 'ZONE'])
prices
```

```python
prices['price'] = pd.to_numeric(prices['PRICE'])
prices = prices.drop(columns=['PRICE'])
prices
```

```python
feb17 = pd.read_csv("feb_2017.csv")
feb17
```

```python
feb17['timestamp'] = feb17['a.local_interval_time']
feb17 = feb17.drop(columns=['a.local_interval_time'])
feb17
```

```python
feb17['timestamp'] = pd.to_datetime(feb17['timestamp'])
feb17['lp_value'] = pd.to_numeric(feb17['a.lp_value'])
feb17['Latitude'] = pd.to_numeric(feb17['b.lat'])
feb17['Longitude'] = pd.to_numeric(feb17['b.long'])
feb17['u_id'] = pd.to_numeric(feb17['b.u_id'])
feb17 = feb17.drop(columns=['b.lat', 'b.long', 'a.lp_value', 'b.u_id'])
feb17
```

```python
feb17_prices = pd.merge(prices, feb17, on='timestamp', how='inner')
feb17_prices
```

```python
feb17_prices = feb17_prices.dropna()
feb17_prices = feb17_prices.drop(columns=['hour', 'minute'])
feb17_prices['lp_price'] = feb17_prices['price']*feb17_prices['lp_value']/100
feb17_prices
```

```python
consumption = feb17_prices.groupby('u_id').sum()
```

```python
consumption = consumption.drop(columns=['price', 'lp_value', 'Longitude', 'Latitude'])
consumption
```

```python
address = pd.read_csv('address.csv')
address
```

```python
address.set_index('u_id', inplace=True)
```

```python
address
```

```python
address.columns
```

```python
house_consumption = pd.merge(consumption, address, on=['u_id'], how='inner')
house_consumption
```

```python
house_consumption = house_consumption.drop(columns=['Unnamed: 0'])
house_consumption
```

```python
house_consumption.to_csv('feb17_house_consumption.csv')
```
